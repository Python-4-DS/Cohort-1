```python
# %maptplotlib magic command to enable interactive plots in notebooks
# inline attr to enable static pplots within the notebook
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# set default plot style
plt.style.use('seaborn-whitegrid')
```

#### all plots start by creating a figure and an axes.

```python
# Figure
fig = plt.figure()

# axes
ax = plt.axes()

# this creates a figure in its simplest for
```

## A simple sinusoid

```python
# The data
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))
```

```python
# Multiple lines
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

## Labelled axes with custom colors and legend

```python
# Plot fn with x data, ydata, line color(and style) and label
# Sinusoid
plt.plot(x, np.sin(x), color='green', label='sin x', linestyle='solid')

# Cosine
plt.plot(x, np.cos(x), color='blue', linestyle='dotted', label='cos x')

# Set title, x-axis label and y-axis label
plt.title('Sine and cosine waves')
plt.xlabel('x')
# plt.axis('equal')

# Set y-limits
plt.ylim(-2, 2)

# Include a legend at the lower-center
plt.legend(loc='lower center', ncol=2)
```
